```python
gb = df.groupby("ParentEduc").agg({'MathScore':'mean',"ReadingScore" : 'mean','WritingScore':'mean'})
print(gb)
```

```python
plt.figure(figsize=(8,4))
sns.heatmap(gb,annot=True)
plt.title('Relationship between Parents Education and Student Score')
plt.show()
```

```python
# from the above chart we have concluded that the education of parents have a good impact on there scores
```

```python
gb_1 = df.groupby("ParentMaritalStatus").agg({'MathScore':'mean',"ReadingScore" : 'mean','WritingScore':'mean'})
print(gb_1)
```

### ðŸ’¡ Distribution od Ethnic Group

```python
import matplotlib.pyplot as plt

# Count how many rows belong to each group
groupA = df.loc[df['EthnicGroup'] == "group A"].count()
groupB = df.loc[df['EthnicGroup'] == "group B"].count()
groupC = df.loc[df['EthnicGroup'] == "group C"].count()
groupD = df.loc[df['EthnicGroup'] == "group D"].count()
groupE = df.loc[df['EthnicGroup'] == "group E"].count()

# Labels
l = ['group A', 'group B', 'group C', 'group D', 'group E']

# Extract counts for the EthnicGroup column
mlist = [
    groupA['EthnicGroup'],
    groupB['EthnicGroup'],
    groupC['EthnicGroup'],
    groupD['EthnicGroup'],
    groupE['EthnicGroup']
]

# Create pie chart
plt.pie(mlist, labels=l, autopct='%1.1f%%', startangle=90)
plt.title('Ethnic Group Distribution')
plt.show()
```
